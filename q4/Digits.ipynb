```python
#Problem Number 1

#Set Uo
import matplotlib.pyplot as plt
import numpy as np
mu, sigma = 3, 3 # mean and standard deviation
mu2,sigma2 = 4, 2
X1 = np.random.normal(mu, sigma, 100)
X2 = np.random.normal(mu, sigma, 100)+np.random.normal(mu2, sigma2, 100)

#Part A

x1_mean=sum(X1)/len(X1)
x2_mean=sum(X2)/len(X2)
print "Mean of X1",x1_mean
print "Mean of X2",x2_mean

#Part B
print "Covariance Matrix"
covariance_matrix=np.cov([X1,X2])
print covariance_matrix

#Part C
w,v=np.linalg.eig(covariance_matrix)
print "Eigen vectors"
print v
print "Eigen values"
print w

#Part D.i
plt.figure()
plt.plot(X1,X2,"ro")
plt.xlabel("X1")
plt.ylabel("X2")
plt.axis('equal')
plt.axis([-20,20,-15,15])
    
#Part D.ii
eigV=np.array([[x1_mean,x2_mean,w[0]*v[0][0], w[0]*v[1][0]],[x1_mean,x2_mean,w[1]*v[0][1],w[1]*v[1][1]]])
X,Y,U,V = zip(*eigV)
ax = plt.gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1)
plt.draw()
plt.show()

#Part E
U=0;
if w[0]>w[1]:
    U=np.array([[v[0][0],v[1][0]],[v[0][1],v[1][1]]])
else:
    U=np.array([[v[0][1],v[1][1]],[v[0][0],v[1][0]]])

centered_x1=[i-x1_mean for i in X1]
centered_x2=[i-x2_mean for i in X2]


rotated=[np.dot(U.T,np.array([centered_x1[i],centered_x2[i]])) for i in xrange(len(centered_x1))]
plt.figure()
plt.plot(*zip(*rotated), marker='o', color='r', ls='')
w,v=np.linalg.eig(np.cov(*zip(*rotated)))
eigV=np.array([[0,0,w[0]*v[0][0], w[0]*v[1][0]],[0,0,w[1]*v[0][1],w[1]*v[1][1]]])
X,Y,U,V = zip(*eigV)
ax = plt.gca()
ax.quiver(X,Y,U,V,angles='xy',scale_units='xy',scale=1)

plt.xlabel("X1")
plt.ylabel("X2")
plt.axis([-20,20,-15,15])
plt.show()


```

```python
#Set up covariance and mean for Method 1

import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
from operator import sub

train=scipy.io.loadmat("../data/digit-dataset/train.mat")
test=scipy.io.loadmat("../data/digit-dataset/test.mat")

labels=train["train_label"]
images=train["train_image"]

images=images.T
images=[images[i].T for i in xrange(len(images))]
images=[np.reshape(images[i].T,(784)) for i in xrange(len(images))]
norms=np.linalg.norm(np.array(images), ord=2, axis=1)
images=[images[i]/norms[i] for i in xrange(len(norms))] 


###Split images into classes
class_images=[[] for i in xrange(10)]
for i in xrange(len(images)):
    class_images[labels[i][0]].append(images[i])


total=[[0]*len(images[0]) for i in xrange(10)]
    
for number in xrange(10):
    for img in class_images[number]:
        total[number]+=img

mean=[total[number]*1.0/len(class_images[number]) for number in xrange(10)]
covariance=[np.cov(class_images[number], rowvar=0) for number in xrange(10)]
overall_cov=np.cov(images,rowvar=0);


new_cov=overall_cov+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[0]=covariance[0]+np.matrix(np.identity(overall_cov.shape[0]))*1 #-3,-1,-4,-3,-3,-3,-3,-2,-2,-3,-2
covariance[1]=covariance[1]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[2]=covariance[2]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[3]=covariance[3]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[4]=covariance[4]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[5]=covariance[5]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[6]=covariance[6]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[7]=covariance[7]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[8]=covariance[8]+np.matrix(np.identity(overall_cov.shape[0]))*1
covariance[9]=covariance[9]+np.matrix(np.identity(overall_cov.shape[0]))*1


for i in xrange(10):
    new_cov=overall_cov+np.matrix(np.identity(overall_cov.shape[0]))*i/10
    print "for i",i, "determininant is", np.linalg.det(new_cov)



#for i in xrange(len(covariance)):
#    covariance[i]=covariance[i]+np.matrix(np.identity(overall_cov.shape[0]))*pow(10,-5)

plt.matshow(covariance[0])
plt.title('Covariance matrix matrix')
plt.colorbar()
plt.show()    

print "The covariance determint is", np.linalg.det(new_cov)
```
